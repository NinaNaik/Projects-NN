# CORRELATION MATRIX

```python
# Correlation Matrix
corr_matrix=df[['Advertising','Sales Revenue','Temperature','Competitors']].corr()
print(corr_matrix)
```

```python
# Advertising and sales has a strong correlation of 99%, Advertising helps to increase Sales Revenue
# Sales Revenue & temperature has a negative correlation
# Competitors have a strong correlation with Advertising and Sales Revenue, Competitors help to increase Sales Revenue in this data/case
# Temperature and advertising have a negative correlation, 
# Temperature and competitors dont have any correlation
```

# HEATMAP

```python
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f', linewidths=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# lINEAR REGRESSION (ADVERTISING AND SALES REVENUE)
from sklearn.linear_model import LinearRegression
```

```python
# NOTE: linearRegression is one word
```

```python
# Reshape data for fitting
X=df[['Advertising']].values.reshape(-1,1)
y=df['Sales Revenue'].values
```

```python
# Creating and fitting the model
model=LinearRegression()
model.fit(X,y)
```

```python
# Predict the sales revenue based on Advertising
```

```python
df['predicted Sales Revenue']=model.predict(X)
```

```python
# Plot the Regression Line
```

```python
plt.scatter(df['Advertising'],df['Sales Revenue'],color='blue')
plt.plot(df['Advertising'],df['Sales Revenue'],color='red')
plt.title('Advertising vs Sales Revenue (Linear Regression)')
plt.xlabel('Advertising ($)')
plt.ylabel('Sales Revenue($)')
plt.show()
```

```python
# The higher the advertisment spend, higher is the sales revenue
```

```python
# Print Model Co-efficients
```

```python
print(f"Intercept:{model.intercept_}")
```

```python
print(f"Slope(Coeficient):{model.coef_[0]}")
```
