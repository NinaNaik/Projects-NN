```python
# REGRESSION MODELS
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.preprocessing import StandardScaler
```

# FEATURE SCALING & SEPARATING TARGET VARIABLE

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
# feature scaling
```

```python
scaler=StandardScaler()
X_Scaled=scaler.fit_transform(X)
```

# TRAIN TEST SPLIT

```python
# Split data into training and testing sets
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1111)
```

# INITIALIZE THE LINEAR REGRESSION MODEL

```python
lr=LinearRegression()
```

# Train the Model

```python
lr.fit(X_train,y_train)
```

# PREDICTION

```python
y_pred=lr.predict(X_test)
y_pred
```

# EVALUATE THE MODEL

```python
mse_lr=mean_squared_error(y_test,y_pred)
r2_lr=r2_score(y_test,y_pred)
print(f'Linear Regression - MSE:{mse_lr},R2:{r2_lr}')
```

```python
# NOTE: THE PERFORMANCE OF THIS MODEL IS 73.35% 
```

# RANDOM FORREST REGRESSOR

```python
# INITIALIZE THE RANDOM FORREST REGRESSOR MODEL
from sklearn.ensemble import RandomForestClassifier
dt_model=RandomForestClassifier()
dt_model.fit(X_train, y_train)
y_pred = dt_model.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
rf_model=RandomForestRegressor(n_estimators=100,random_state=42)
rf_model
```

# TRAIN THE MODEL

```python
rf_model.fit(X_train,y_train)
```

# PREDICTIONS

```python
y_pred=rf_model.predict(X_test)
y_pred
```

# EVALUATE THE MODEL

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
mse_rf=mean_squared_error(y_test,y_pred)
mse_rf
```

```python
r2_rf=r2_score(y_test,y_pred)
print(f'RandomForest-MSE:{mse_rf},R2:{r2_rf}')
```

```python
# NOTE: r2 score of Random Forrest Model is 69% and is a much better model thanthe Decision Tree Model which has a r2 score of 25% only
# LINEAR REGRESSION R2 SCORE IS HIGHEST AT 73% and better than the other two models
```
